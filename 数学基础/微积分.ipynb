# 导数，偏导，梯度

### ✔ 一元求导（普通导数）

**“这条线在这个点的斜率是多少？”**

对于一元函数 $f(x)$，其导数定义为：
$$
f'(x) = \lim_{\Delta x \to 0} \frac{f(x + \Delta x) - f(x)}{\Delta x}
$$

### ✔ 多元求导（偏导数）

**偏导 = 沿某一个方向的斜率（方向导数的一种），把其他参数当成一个常数**

**“我只沿着 $x_1$ 方向走一点点，$L$ 上升还是下降？变化有多快？”**

对于多元函数 $f(x_1, x_2, \dots, x_n)$，偏导数定义为：
$$
\frac{\partial f}{\partial x_i} = \lim_{\Delta x_i \to 0} \frac{f(x_1, \dots, x_i + \Delta x_i, \dots, x_n) - f(x_1, \dots, x_i, \dots, x_n)}{\Delta x_i}
$$

你可以把它想象成“多维地形”：

- 导数 = 山坡的一段斜度
- 偏导 = 向东走的斜度、向北走的斜度、向上走的斜度

对于导数和偏导而言，有两个比较重要的性质：
1. 导数为 0 的时候，说明斜率不变化了，所以要么是最低点，要么是最高点，要么是一个鞍点。
2. 导数的方向代表着变化率最快的方向，这也是梯度的基本支持。

### ✔ 梯度

梯度是由所有偏导数组成的向量，这个向量代表着函数最快上升的方向。原因是函数最快上升的方向（变化率最大的方向）就是偏导的方向。

对于多元函数 $f(x_1, x_2, \dots, x_n)$，梯度定义为：
$$
\nabla f = \left( \frac{\partial f}{\partial x_1}, \frac{\partial f}{\partial x_2}, \dots, \frac{\partial f}{\partial x_n} \right)
$$

梯度是一个向量，是指明方向的。

### 梯度下降

负梯度的方向，就是最快下降的方向。因为我们常常用多元方程充当损失函数（Loss Function），而机器学习领域大部分情况下，都是要使得损失函数最小，因此常常采用梯度下降法，来快速逼近求解空间内的最低点所对应的参数。

梯度下降的更新公式为：
$$
\theta^{(t+1)} = \theta^{(t)} - \alpha \nabla f(\theta^{(t)})
$$
其中：
- $\theta$ 是参数向量
- $\alpha$ 是学习率（步长）
- $\nabla f(\theta^{(t)})$ 是当前参数的梯度

梯度下降的核心思想是：沿着负梯度方向（即最快下降方向）更新参数，直到找到局部最小值。
